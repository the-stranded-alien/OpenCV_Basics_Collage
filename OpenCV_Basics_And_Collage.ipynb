```python
import cv2 # OpenCV
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Read and Display an Image using Python

img = cv2.imread("Images/b1.png")
print(img.shape)

# In OpenCV order followed is BGR whereas in Matplotlib the order followed is RGB.
plt.imshow(img)
plt.show()
```

```python
# Convert from BGR to RGB Color Space
img = cv2.cvtColor(img, 4) # COLOR_BGR2RGB => Code == 4
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
image_names = os.listdir("Images")

for f in image_names:
    if(f.endswith(".png")):
        file_path = "Images/" + f
        img = cv2.imread(file_path)
        img = cv2.cvtColor(img, 4)
        plt.imshow(img)
        plt.title(img.shape)
        plt.show()
```

```python
image_names = os.listdir("Images")

for f in image_names:
    if(f.endswith(".png")):
        file_path = "Images/" + f
        img = cv2.imread(file_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        plt.imshow(img, cmap='gray')
        plt.title(img.shape)
        plt.show()
```

# Image Collage !!

```python
red = np.zeros((50,50,3), dtype='int')
red[:,:,0] = 180
green = np.zeros((50,50,3), dtype='int')
green[:,:,1] = 170
plt.imshow(red)
plt.show()
plt.imshow(green)
plt.show()
```

```python
red_green = np.hstack((red, green))
green_red = red_green[ : , : : -1, : ]
img = np.vstack((red_green, green_red))
plt.imshow(img)
plt.show()
```

```python
image_names = os.listdir("Images")

animals = []
for f in image_names:
    if(f.endswith(".png")):
        file_path = "Images/" + f
        img = cv2.imread(file_path)
        img = cv2.cvtColor(img, 4)
        img = cv2.resize(img, (100,100))
        animals.append(img)

# Make A Collage
border_top = np.zeros((5,215,3),dtype='uint8')
border_bottom = np.zeros((5,215,3),dtype='uint8')
border_right = np.zeros((205,5,3),dtype='uint8')
border_left =  np.zeros((205,5,3),dtype='uint8')
partition_vertical = np.zeros((5,205,3),dtype='uint8')
partition_horizontal = np.zeros((100,5,3),dtype='uint8')
animal_12 = np.hstack((animals[0], partition_horizontal, animals[1]))
animal_34 = np.hstack((animals[2], partition_horizontal, animals[3]))
collage = np.vstack((animal_12, partition_vertical, animal_34))
collage = np.hstack((border_left, collage, border_right))
collage = np.vstack((border_top, collage, border_bottom))
plt.imshow(collage)
plt.title(collage.shape)
plt.show()
```

```python
fun_collage_1 = np.hstack((collage, collage))
fun_collage_2 = fun_collage_1[ : , : : -1, : ]
fun_collage = np.vstack((fun_collage_1, fun_collage_2))
plt.figure(figsize=(10,10))
plt.imshow(fun_collage)
plt.title(fun_collage.shape)
plt.show()
```
